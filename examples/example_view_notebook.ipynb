# gleplot view() Function Demo

This notebook demonstrates the `view()` function for displaying plots inline in Jupyter notebooks.

```python
import sys
import numpy as np
from pathlib import Path

# Add src to path
sys.path.insert(0, str(Path.cwd().parent / 'src'))

import gleplot as glp
```

## Example 1: Basic Line Plot with view()

```python
# Create figure and plot
fig = glp.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

# Create data
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot
ax.plot(x, y_sin, color='blue', label='sin(x)', linestyle='-')
ax.plot(x, y_cos, color='red', label='cos(x)', linestyle='--')

ax.set_xlabel('x (radians)')
ax.set_ylabel('y')
ax.set_title('Sine and Cosine Functions')
ax.legend()

# Display inline
fig.view()
```

## Example 2: Scatter Plot with Custom DPI

```python
# Create figure
fig = glp.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

# Random data
np.random.seed(42)
x = np.random.randn(50)
y = 2*x + np.random.randn(50) * 0.5

# Scatter plot
ax.scatter(x, y, color='purple', s=20, marker='o', label='Data')

# Trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
x_line = np.linspace(x.min(), x.max(), 100)
ax.plot(x_line, p(x_line), color='red', linestyle='--', label='Trend')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot with Trend Line (High DPI)')
ax.legend()

# Display with high DPI for better quality
fig.view(dpi=200)
```

## Example 3: Using Module-Level Functions

```python
# Use matplotlib-style API
fig, ax = glp.subplots(figsize=(8, 6))

# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]
x_pos = np.arange(len(categories))

ax.bar(x_pos, values, color='green', label='Values')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Chart Example')

# Use module-level view function
glp.view()
```

## Example 4: Subplots Grid

```python
# Create 2x2 subplot grid
fig, axes = glp.subplots(2, 2, figsize=(12, 10))

# Top-left: Line plot
x = np.linspace(0, 10, 50)
axes[0].plot(x, np.exp(-x/5), 'b-', label='exponential decay')
axes[0].set_title('Exponential Decay')
axes[0].legend()

# Top-right: Scatter
axes[1].scatter(np.random.randn(30), np.random.randn(30), color='red')
axes[1].set_title('Random Scatter')

# Bottom-left: Bar
axes[2].bar([1, 2, 3, 4], [10, 20, 15, 25], color='green')
axes[2].set_title('Bar Chart')

# Bottom-right: Multiple lines
x = np.linspace(0, 5, 50)
axes[3].plot(x, x, 'r-', label='linear')
axes[3].plot(x, x**2, 'b-', label='quadratic')
axes[3].set_title('Polynomials')
axes[3].legend()

# Display the entire subplot grid
fig.view()
```

## Notes

- The `view()` function automatically detects if running in a Jupyter notebook
- In notebooks, plots are displayed inline using IPython.display
- In regular Python scripts, the function saves to a temporary file
- You can customize the DPI for better resolution: `fig.view(dpi=300)`
- The function requires GLE to be installed for compilation
